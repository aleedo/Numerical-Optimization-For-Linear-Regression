# Numerical Optimization For Linear Regression 

```python
from LinearRegressionOptimization import LinearRegression
from LinearRegressionOptimization.input import get_from_csv
from LinearRegressionOptimization.optimizers import *
```

```python
X, y = get_from_csv()
```

## SGD

```python
full_batch_gd = LinearRegression(
    batch_size=X.shape[0],
    max_epochs=300,
    optimizer=Sgd(learning_rate=0.0001),
    random_state=42,
    cost_threshold=0.01,
    gradient_threshold=0.01,
    normalize=False,
)

full_batch_gd.fit(X, y)
```


### Training Report

```python
print(full_batch_gd.training_report())
```

### Evaluation

```python
y_pred = full_batch_gd.predict(X)
print(full_batch_gd.evaluate(y, y_pred))
```

### Training Visualization

#### Training Loss

```python
full_batch_gd.plot_loss()
```

#### Parameters Loss

```python
full_batch_gd.plot_theta_loss()
```

#### Regression Lines

```python
full_batch_gd.plot_regression_lines()
```

#### Best Fit Regression Line

```python
full_batch_gd.plot_best_fit()
```
